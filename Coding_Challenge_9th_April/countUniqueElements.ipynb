```python
# A report counts unique items
#---------------------------------------------
# Function to count unique elements in a list
def count_unique(lst):
    # Initialize empty list to store unique elements
    unique_list = []

    # Loop through each element
    for item in lst:
        # Check if item is not already in unique_list
        if item not in unique_list:
            unique_list.append(item)   # Add unique item

    # Return count of unique elements
    return len(unique_list)


# Taking input (example list)
lst = [1, 1, 2, 3, 3]

# Calling the function and storing result
result = count_unique(lst)

# Printing the result
print(result)
```
