```python
# A text processing tool identifies vowels
#---------------------------------------------------
# Function to check whether a character is a vowel
def check_vowel(ch):
    # Check if the character is in vowel list (both lowercase and uppercase)
    if ch in "aeiouAEIOU":
        return "Vowel"
    else:
        return "Not Vowel"


# Taking input from user
ch = input("Enter a character: ")

# Calling the function and storing result
result = check_vowel(ch)

# Printing the result
print(result)
```
