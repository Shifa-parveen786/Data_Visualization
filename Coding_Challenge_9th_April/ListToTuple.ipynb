```python
# A converter changes data structure
#-------------------------------------------------------
# Function to convert list to tuple
def list_to_tuple(lst):
    # Convert list to tuple using tuple() constructor
    result = tuple(lst)

    # Return tuple
    return result


# Taking input (example list)
lst = [1, 2, 3]

# Calling the function and storing result
result = list_to_tuple(lst)

# Printing the result
print(result)
```
