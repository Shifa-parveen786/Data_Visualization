```python
# A system stores student marks
#-----------------------------------------
# Function to find student with highest marks
def highest_marks(students):
    # Get first student name
    top_student = list(students.keys())[0]

    # Assume first student's marks as highest
    max_marks = students[top_student]

    # Loop through dictionary
    for name in students:
        # Check if current student's marks are higher
        if students[name] > max_marks:
            max_marks = students[name]   # Update max marks
            top_student = name           # Update student name

    # Return student with highest marks
    return top_student


# Taking input (example dictionary)
students = {"A": 80, "B": 95, "C": 78}

# Calling the function
result = highest_marks(students)

# Printing the result
print(result)
```
