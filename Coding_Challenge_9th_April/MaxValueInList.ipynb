```python
# A data processor finds maximum value
#---------------------------------------------------
# Function to find maximum value in a list (without using built-in functions)
def find_max(numbers):
    # Assume first element is the maximum
    maximum = numbers[0]

    # Loop through the list
    for num in numbers:
        # Check if current number is greater than maximum
        if num > maximum:
            maximum = num   # Update maximum

    # Return the maximum value
    return maximum


# Taking input (example list)
numbers = [5, 8, 2]

# Calling the function and storing result
result = find_max(numbers)

# Printing the result
print(result)
```
