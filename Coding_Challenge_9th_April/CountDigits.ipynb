```python
# A data processor finds maximum values
#----------------------------------------------
# Function to count number of digits in a number
def count_digits(num):
    # Initialize counter
    count = 0

    # Loop until number becomes 0
    while num > 0:
        count += 1      # Increase count
        num //= 10      # Remove last digit

    # Return total count
    return count


# Taking input from user
num = int(input("Enter number: "))

# Calling the function and storing result
result = count_digits(num)

# Printing the result
print(result)
```
