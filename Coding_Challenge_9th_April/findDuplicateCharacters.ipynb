```python
# A data cleaner identifies duplicats
#------------------------------------------------
# Function to find duplicate characters in a string
def find_duplicates(text):
    # Dictionary to store frequency of characters
    freq = {}

    # Count frequency of each character
    for ch in text:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1

    # Initialize result string
    result = ""

    # Find characters with frequency greater than 1
    for ch in freq:
        if freq[ch] > 1:
            result += ch + " "   # Add duplicate character

    # Return result (strip to remove extra space)
    return result.strip()


# Taking input from user
text = input("Enter a string: ")

# Calling the function and storing result
result = find_duplicates(text)

# Printing the result
print(result)
```
