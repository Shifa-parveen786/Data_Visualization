```python
# A ranking system finds second top score
#------------------------------------------
# Function to find second largest element in a list
def second_largest(lst):
    # Assume first element as largest and second largest
    largest = second = lst[0]

    # Loop through the list
    for num in lst:
        # Update largest and second largest
        if num > largest:
            second = largest
            largest = num
        elif num > second and num != largest:
            second = num

    # Return second largest element
    return second


# Taking input (example list)
lst = [10, 20, 5, 15]

# Calling the function and storing result
result = second_largest(lst)

# Printing the result
print(result)
```
