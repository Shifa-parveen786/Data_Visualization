```python
# A system removes duplicate enteries
#-------------------------------------------------------
# Function to remove duplicates from a list
def remove_duplicates(lst):
    # Initialize empty list to store unique elements
    unique_list = []

    # Loop through each element in the list
    for item in lst:
        # Check if item is not already in unique_list
        if item not in unique_list:
            unique_list.append(item)   # Add unique item

    # Return list without duplicates
    return unique_list


# Taking input (example list)
lst = [1, 1, 2, 3]

# Calling the function and storing result
result = remove_duplicates(lst)

# Printing the result
print(result)
```
