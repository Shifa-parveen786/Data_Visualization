```python
# a sequence is missing a number
#----------------------------------------------------
# Function to find missing number in a sequence
def find_missing_number(lst):
    # Find maximum number in list manually
    max_num = lst[0]
    for num in lst:
        if num > max_num:
            max_num = num

    # Calculate expected sum from 1 to max_num
    expected_sum = (max_num * (max_num + 1)) // 2

    # Calculate actual sum of list
    actual_sum = 0
    for num in lst:
        actual_sum += num

    # Missing number = expected - actual
    missing = expected_sum - actual_sum

    return missing


# Taking input (example list)
lst = [1, 2, 4, 5]

# Calling the function
result = find_missing_number(lst)

# Printing the result
print(result)
```
