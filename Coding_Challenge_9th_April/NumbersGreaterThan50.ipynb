```python
# A report counts high-values enteries
#------------------------------------------------
# Function to count numbers greater than 50 in a list
def count_greater_than_50(numbers):
    # Initialize counter
    count = 0

    # Loop through each number in the list
    for num in numbers:
        # Check if number is greater than 50
        if num > 50:
            count += 1   # Increase count

    # Return total count
    return count


# Taking input (example list)
numbers = [10, 60, 30, 80]

# Calling the function and storing result
result = count_greater_than_50(numbers)

# Printing the result
print(result)
```
