```python
# A System sorts data manually
#----------------------------------------------
# Function to sort a list without using built-in sort()
def sort_list(lst):
    # Get length of list
    n = len(lst)

    # Bubble Sort Algorithm
    for i in range(n):
        for j in range(0, n - i - 1):
            # Compare adjacent elements
            if lst[j] > lst[j + 1]:
                # Swap if elements are in wrong order
                lst[j], lst[j + 1] = lst[j + 1], lst[j]

    # Return sorted list
    return lst


# Taking input (example list)
lst = [3, 1, 2]

# Calling the function and storing result
result = sort_list(lst)

# Printing the result
print(result)
```
