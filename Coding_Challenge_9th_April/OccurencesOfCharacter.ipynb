```python
# A system tracks character frequency
#---------------------------------------
# Function to count occurrences of a character in a string
def count_character(text, ch):
    # Initialize counter
    count = 0

    # Loop through each character in the string
    for c in text:
        # Check if current character matches given character
        if c == ch:
            count += 1   # Increase count

    # Return total count
    return count


# Taking input from user
text = input("Enter a string: ")
ch = input("Enter character to count: ")

# Calling the function and storing result
result = count_character(text, ch)

# Printing the result
print(result)
```
