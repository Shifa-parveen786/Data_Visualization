```python
# A store tracks daily sales
#---------------------------------------------------
# Function to calculate total weekly sales
def total_sales(sales_list):
    # Initialize total to 0
    total = 0

    # Loop through each day's sale
    for sale in sales_list:
        total += sale   # Add each sale to total

    # Return the final total
    return total


# Taking input from user (example list)
sales_list = [100, 200, 150, 300, 250, 400, 100]

# Calling the function and storing result
result = total_sales(sales_list)

# Printing the result
print(result)
```
