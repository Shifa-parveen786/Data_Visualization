```python
# An online platform categories users by age group
#----------------------------------------------------
# Function to classify age group
def classify_age(age):
    # Check if age is less than 18
    if age < 18:
        return "Minor"

    # Check if age is between 18 and 59
    elif age < 60:
        return "Adult"

    # If age is 60 or above
    else:
        return "Senior"


# Taking input from user
age = int(input("Enter age: "))

# Calling the function and storing result
result = classify_age(age)

# Printing the result
print(result)
```
