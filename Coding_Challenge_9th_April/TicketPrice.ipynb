```python
# A ticket booking system calculates ticket price based on day.
#---------------------------------------------------------------------
# Function to determine ticket price based on day
def get_ticket_price(day):
    # Check if the day is Sunday
    if day == "Sunday":
        return 200

    # For all other days
    else:
        return 100


# Taking input from user
day = input("Enter day: ")

# Calling the function and storing result
result = get_ticket_price(day)

# Printing the result
print(result)
```
