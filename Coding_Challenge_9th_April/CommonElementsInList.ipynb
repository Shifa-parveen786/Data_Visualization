```python
# A tool finds common users
#-------------------------------------------------------
# Function to find common elements between two lists
def find_common(list1, list2):
    # Initialize empty list to store common elements
    common = []

    # Loop through first list
    for item in list1:
        # Check if item is present in second list and not already added
        if item in list2 and item not in common:
            common.append(item)   # Add common item

    # Return list of common elements
    return common


# Taking input (example lists)
list1 = [1, 2, 3]
list2 = [2, 3, 4]

# Calling the function and storing result
result = find_common(list1, list2)

# Printing the result
print(result)
```
