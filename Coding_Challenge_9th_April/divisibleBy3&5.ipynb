```python
# A payment system process only numbers divisible by both 3 and 5
#-------------------------------------------------------------
# Function to check if number is divisible by both 3 and 5
def check_divisible(num):
    # Check condition using modulus operator
    if num % 3 == 0 and num % 5 == 0:
        return "Yes"
    else:
        return "No"


# Taking input from user
num = int(input("Enter number: "))

# Calling the function and storing result
result = check_divisible(num)

# Printing the result
print(result)
```
