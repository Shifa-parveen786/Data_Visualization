```python
# A masking tool hide vowels
#-------------------------------------------------
# Function to replace vowels with '*'
def mask_vowels(text):
    # Initialize empty string to store result
    result = ""

    # Loop through each character in the string
    for ch in text:
        # Check if character is a vowel
        if ch in "aeiouAEIOU":
            result += "*"   # Replace vowel with '*'
        else:
            result += ch    # Keep character as it is

    # Return modified string
    return result


# Taking input from user
text = input("Enter a string: ")

# Calling the function and storing result
result = mask_vowels(text)

# Printing the result
print(result)
```
