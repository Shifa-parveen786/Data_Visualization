```python
# A formatter converts text
#-----------------------------------------------------
# Function to convert string to uppercase without using built-in methods
def to_uppercase(text):
    # Initialize empty string to store result
    result = ""

    # Loop through each character in the string
    for ch in text:
        # Check if character is lowercase letter
        if 'a' <= ch <= 'z':
            # Convert to uppercase using ASCII values
            result += chr(ord(ch) - 32)
        else:
            # Keep character as it is (for spaces, symbols, etc.)
            result += ch

    # Return uppercase string
    return result


# Taking input from user
text = input("Enter a string: ")

# Calling the function and storing result
result = to_uppercase(text)

# Printing the result
print(result)
```
