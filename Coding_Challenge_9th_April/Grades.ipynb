```python
# An exam system assigns grades
#--------------------------------------------------
# Function to assign grade based on marks
def assign_grade(marks):
    # Check if marks are 90 or above
    if marks >= 90:
        return "A"

    # Check if marks are between 75 and 89
    elif marks >= 75:
        return "B"

    # Check if marks are between 50 and 74
    elif marks >= 50:
        return "C"

    # If marks are below 50
    else:
        return "Fail"


# Taking input from user
marks = int(input("Enter marks: "))

# Calling the function and storing result
result = assign_grade(marks)

# Printing the result
print(result)
```
