```python
#  A system rotates list data
#--------------------------------------
# Function to rotate list by one position (right rotation)
def rotate_list(lst):
    # Check if list is not empty
    if len(lst) == 0:
        return lst

    # Store last element
    last = lst[-1]

    # Shift elements to the right
    for i in range(len(lst) - 1, 0, -1):
        lst[i] = lst[i - 1]

    # Place last element at first position
    lst[0] = last

    # Return rotated list
    return lst


# Taking input (example list)
lst = [1, 2, 3]

# Calling the function and storing result
result = rotate_list(lst)

# Printing the result
print(result)
```
