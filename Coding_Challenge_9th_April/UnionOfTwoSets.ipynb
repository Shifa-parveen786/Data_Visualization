```python
# A system merges sets
#----------------------------------------------------------------
# Function to find union of two sets
def find_union(set1, set2):
    # Initialize result set with elements of set1
    result = set1.copy()

    # Add elements of set2 into result
    for item in set2:
        result.add(item)   # Set automatically handles duplicates

    # Return union set
    return result


# Taking input (example sets)
set1 = {1, 2}
set2 = {2, 3}

# Calling the function and storing result
result = find_union(set1, set2)

# Printing the result
print(result)
```
