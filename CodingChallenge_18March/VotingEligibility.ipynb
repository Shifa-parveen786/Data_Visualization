```python
# check eligible for vote or not
#------------------------------------------------------------
#with input validation
while True:
    age = int(input("Enter your age: "))
    if age < 0:
        print("Invalid age! Please enter a positive number.")
    else:
        break

if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```
