```python
# printing even numbers
#-------------------------------------------------
# using recursion
def print_even(n):
    if n < 2:
        return
    print_even(n - 2)
    print(n)

n = int(input("Enter a number: "))
if n % 2 != 0:
    n -= 1

print_even(n)
```
