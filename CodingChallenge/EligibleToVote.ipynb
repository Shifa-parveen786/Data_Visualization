```python
# Check whether a person is eligible to vote or not
#---------------------------------------------------
# Defining the function
def check_vote(age):
    if age >= 18:
        return "Eligible to vote"
    else:
        return "Not eligible to vote"
#------------------------------------------------------
age = int(input("Enter your age: "))
#------------------------------------------------------
# Calling the function
print(check_vote(age))
```
