```python
# calculate electricity bill based on units consumed
#-------------------------------------------------------
units = int(input("Enter units consumed: "))
bill = 0

remaining = units
# using select statements
#----------------------------------------------------------
if remaining > 200:
    bill += (remaining - 200) * 10
    remaining = 200

if remaining > 100:
    bill += (remaining - 100) * 7
    remaining = 100

bill += remaining * 5

print("Electricity bill: ₹", bill)
```
