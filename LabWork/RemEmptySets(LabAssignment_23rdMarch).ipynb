```python
# Remove empty sets from lists problem statement
#----------------------------------------------------
# Input list of sets
lst = [{1, 2}, set(), {3, 4}, set()]

result = []  # list to store non-empty sets

# Loop through list
for i in lst:
    # Empty set evaluates to False
    if i:
        result.append(i)

# Output result
print("After Removal:", result)
```
