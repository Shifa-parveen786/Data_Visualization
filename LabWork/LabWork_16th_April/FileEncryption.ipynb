```python
f = open("file1.txt", "r")
text = f.read()
f.close()

key = 3
encrypted = ""

for ch in text:
    encrypted += chr(ord(ch) + key)

out = open("encrypted.txt", "w")
out.write(encrypted)
out.close()
```
