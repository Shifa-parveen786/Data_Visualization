```python
# Program to count frequency of each word (ignore case & punctuation)

import string

# open file
file = open("article.txt", "r")

# read content
text = file.read()

# close file
file.close()

# convert to lowercase
text = text.lower()

# remove punctuation
for ch in string.punctuation:
    text = text.replace(ch, "")

# split into words
words = text.split()

# dictionary for counting
freq = {}

for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

# display result
for word in freq:
    print(word, ":", freq[word])
```
