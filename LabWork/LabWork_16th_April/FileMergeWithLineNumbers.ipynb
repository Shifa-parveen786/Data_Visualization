```python
f1 = open("file1.txt", "r")
f2 = open("file2.txt", "r")
out = open("merged.txt", "w")

lines = f1.readlines() + f2.readlines()

for i in range(len(lines)):
    out.write(str(i+1) + ". " + lines[i])

f1.close()
f2.close()
out.close()
```
