```python
f1 = open("file1.txt", "r")
f2 = open("file2.txt", "r")

lines1 = f1.readlines()
lines2 = f2.readlines()

for line in lines1:
    if line not in lines2:
        print(line.strip())

f1.close()
f2.close()
```
