```python
# Program to filter students with marks > 75

# open input file
with open("students.txt", "r") as file:

    # open output file
    with open("top_students.txt", "w") as output:

        # read each line
        for line in file:
            # remove extra spaces/newline
            line = line.strip()

            # split data
            name, marks, city = line.split(",")

            # convert marks to integer
            marks = int(marks)

            # check condition
            if marks > 75:
                # write to new file
                output.write(line + "\n")

print("Filtered data saved successfully!")
```
