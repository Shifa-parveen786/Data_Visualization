```python
# Program to remove duplicate lines and preserve order

# set to store seen lines
seen = set()

# open input file
infile = open("input.txt", "r")

# open output file
outfile = open("output.txt", "w")

# read line by line
for line in infile:
    if line not in seen:
        outfile.write(line)
        seen.add(line)

# close files
infile.close()
outfile.close()

print("Duplicate lines removed successfully!")
```
