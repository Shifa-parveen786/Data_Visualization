```python
# Pattern printing with conditions
rows = int(input("Enter number of rows: "))

for i in range(1, rows + 1):
    # Check if row is even or odd
    if i % 2 == 0:
        for j in range(i):
            print("*", end=" ")
    else:
        for j in range(i):
            print("#", end=" ")

    print()  # Move to next line
```
