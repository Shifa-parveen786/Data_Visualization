```python
# Number classification system
def classify_number(n):
    # Positive / Negative / Zero
    if n > 0:
        print("Positive", end=", ")
    elif n < 0:
        print("Negative", end=", ")
    else:
        print("Zero", end=", ")

    # Nested if for Even / Odd
    if n != 0:
        if n % 2 == 0:
            print("Even")
        else:
            print("Odd")

# Taking input from user
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

# Processing all numbers using loop
for num in numbers:
    print(num, "->", end=" ")
    classify_number(num)
```
