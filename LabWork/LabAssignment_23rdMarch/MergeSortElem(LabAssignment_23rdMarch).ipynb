```python
# Merge and sort unique elements problem statement
#-------------------------------------------------
# Input lists
list1 = [3, 1, 2]
list2 = [2, 4, 5]

# Merge both lists
merged = list1 + list2

# Remove duplicates using set
unique = list(set(merged))

# Sort the list
unique.sort()

# Output result
print("Sorted Unique List:", unique)
```
