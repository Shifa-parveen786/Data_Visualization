```python
# tuple pair sum problem statement
#----------------------------------------------------
# Input tuple and target sum
t = (1, 2, 3, 4, 5)
target = 5

pairs = []  # list to store pairs

# Use nested loop to check all pairs
for i in range(len(t)):
    for j in range(i + 1, len(t)):
        # Check if sum equals target
        if t[i] + t[j] == target:
            pairs.append((t[i], t[j]))

# Output result
print("Pairs:", pairs)
```
