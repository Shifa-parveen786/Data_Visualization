```python
# Common Elements in three lists problem statement
#--------------------------------------------------
# Input lists
l1 = [1, 2, 3]
l2 = [2, 3, 4]
l3 = [2, 5, 3]

common = []  # list to store common elements

# Check elements in all three lists
for i in l1:
    if i in l2 and i in l3:
        common.append(i)

# Output result
print("Common Elements:", common)
```
