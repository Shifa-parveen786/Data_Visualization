```python
#Faltten nested list problem statement
#----------------------------------------
# Input nested list
nested = [[1, 2], [3, 4], [5]]

flat = []  # list to store flattened elements

# Loop through each sublist
for sublist in nested:
    # Loop through each element in sublist
    for item in sublist:
        flat.append(item)

# Output result
print("Flattened List:", flat)
```
