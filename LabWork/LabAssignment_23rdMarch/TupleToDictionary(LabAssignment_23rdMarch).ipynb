```python
# tuple to dictionary mapping problem statement
#--------------------------------------------------
# Input tuples
keys = ('a', 'b', 'c')
values = (1, 2, 3)

result = {}  # dictionary to store mapping

# Map keys to values
for i in range(len(keys)):
    result[keys[i]] = values[i]

# Output result
print("Dictionary:", result)
```
