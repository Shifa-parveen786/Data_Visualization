```python
# prime number checker
def is_prime(n):
    # Check for numbers less than or equal to 1
    if n <= 1:
        return False

    # Check divisibility
    for i in range(2, n):
        if n % i == 0:
            return False

    return True

# Taking input from user
num = int(input("Enter a number: "))

# Display result
if is_prime(num):
    print(num, "is a Prime Number")
else:
    print(num, "is Not a Prime Number")
```
