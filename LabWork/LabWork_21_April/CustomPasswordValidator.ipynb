```python
# Custom Password Validator

# custom exception
class InvalidPassword(Exception):
    pass


password = input("Enter password: ")

try:
    # rule 1: minimum length
    if len(password) < 8:
        raise InvalidPassword("Password must be at least 8 characters")

    has_upper = False
    has_lower = False
    has_digit = False

    # check each character using loop
    for ch in password:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True

    # check all conditions
    if not has_upper:
        raise InvalidPassword("Must contain uppercase letter")

    if not has_lower:
        raise InvalidPassword("Must contain lowercase letter")

    if not has_digit:
        raise InvalidPassword("Must contain digit")

    print("Valid Password ✅")

except InvalidPassword as e:
    print("Invalid Password ❌:", e)
```
