```python
# tuple element frequency problem statement
#--------------------------------------------------
# Input tuple
t = (1, 2, 2, 3, 1, 4)

freq = {}  # dictionary to store frequency

# Loop through tuple
for i in t:
    # Increase count if element exists, else initialize
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

# Output result
print("Frequency:", freq)
```
