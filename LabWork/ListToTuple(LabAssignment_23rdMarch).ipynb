```python
# convert list to tuple and remove duplicates problem statement
#----------------------------------------------------------
# Input list
lst = [1, 2, 2, 3, 4, 4]

unique = []  # list to store unique elements

# Remove duplicates manually
for i in lst:
    if i not in unique:
        unique.append(i)

# Convert list to tuple
result = tuple(unique)

# Output result
print("Tuple:", result)
```
