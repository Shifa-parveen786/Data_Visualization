```python
# Student Grade analyzer
def calculate_grade(average):
    if average >= 90:
        return "A"
    elif average >= 75:
        return "B"
    elif average >= 50:
        return "C"
    else:
        return "Fail"

total = 0

# Loop for 5 subjects
for i in range(1, 6):
    marks = int(input(f"Enter marks for subject {i}: "))

    # Validation
    if marks < 0 or marks > 100:
        print("Invalid marks! Please enter between 0 and 100.")
        exit()

    total += marks

# Calculate average
average = total / 5

print("Total Marks:", total)
print("Average Marks:", average)
print("Grade:", calculate_grade(average))
```
