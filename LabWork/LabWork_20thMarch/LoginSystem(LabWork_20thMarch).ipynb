```python
# Login System
def login():
    username = "admin"
    password = "1234"
    attempts = 0

    while attempts < 3:
        u = input("Enter username: ")
        p = input("Enter password: ")

        if u == username and p == password:
            print("Login Successful")
            return
        else:
            print("Invalid credentials")
            attempts += 1

    print("Account Locked")

login()
```
