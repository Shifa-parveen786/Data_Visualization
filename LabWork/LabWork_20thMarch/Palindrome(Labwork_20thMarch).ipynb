```python
# palindrome Checker
def is_palindrome(value):
    # Convert to string to handle both number and text
    value = str(value)

    # Reverse using loop
    reversed_value = ""
    for char in value:
        reversed_value = char + reversed_value

    # Check palindrome
    if value == reversed_value:
        return True
    else:
        return False

# Taking input from user
user_input = input("Enter a number or string: ")

# Display result
if is_palindrome(user_input):
    print("Palindrome")
else:
    print("Not Palindrome")
```
