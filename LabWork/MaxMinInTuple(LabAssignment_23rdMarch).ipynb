```python
# find maximum and minimum in tuple problem statement
#----------------------------------------------------
# Input tuple
t = (5, 1, 8, 3)

# Assume first element as max and min
max_val = t[0]
min_val = t[0]

# Loop through tuple
for i in t:
    if i > max_val:
        max_val = i  # update max
    if i < min_val:
        min_val = i  # update min

# Output result
print("Max =", max_val)
print("Min =", min_val)
```
