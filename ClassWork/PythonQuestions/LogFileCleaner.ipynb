```python
# Log File Cleaner

# open file
try:
    f = open("log.txt", "r")

    lines = f.readlines()   # read all lines
    f.close()

    unique_logs = set()     # set for removing duplicates
    error_logs = []         # list to store ERROR logs
    count = 0               # counter

    for line in lines:
        line = line.strip()     # remove extra spaces

        # add to set (duplicates automatically removed)
        unique_logs.add(line)

    # filter only ERROR logs
    for log in unique_logs:
        if "ERROR" in log:
            error_logs.append(log)
            count += 1

    # display result
    print("ERROR Logs:")
    for e in error_logs:
        print(e)

    print("\nTotal ERROR logs:", count)

except:
    print("File not found!")
```
