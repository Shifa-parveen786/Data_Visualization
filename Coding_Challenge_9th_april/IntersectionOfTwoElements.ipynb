```python
# A tool finds common set elements
#-----------------------------------------------
# Function to find intersection of two sets
def find_intersection(set1, set2):
    # Initialize empty set for result
    result = set()

    # Loop through elements of first set
    for item in set1:
        # Check if element is also in second set
        if item in set2:
            result.add(item)   # Add common element

    # Return intersection set
    return result


# Taking input (example sets)
set1 = {1, 2}
set2 = {2, 3}

# Calling the function and storing result
result = find_intersection(set1, set2)

# Printing the result
print(result)
```
